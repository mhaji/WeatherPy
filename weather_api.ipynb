```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import time
from pprint import pprint
from citipy import citipy
import keys
import random
import seaborn as sns
```

```python
# Create empty list to store city names
cities_name = []

#generate random latitude and longitude points
random_lat_long = []
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)


#find nearest city with location points
for location in lat_lngs:
    city = citipy.nearest_city(location[0],location[1]).city_name
#append cities to list  
    if city not in cities_name:
        cities_name.append(city)
        
    
```

```python
print(cities_name)
```

```python
# Import API key
api_key = keys.api_key
```

```python
# Build URL to query data
url = "http://api.openweathermap.org/data/2.5/weather?"
units = "Imperial"
query_url = f"{url}appid={api_key}&units={units}&q="
```

```python
city_data = []

for city in cities_name:
    response = requests.get(query_url + city).json()
    pprint(response)
    
    if response["cod"] != "404":
        city_lat = response["coord"]["lat"]
        city_lng = response["coord"]["lon"]
        city_temp = response["main"]["temp"]
        city_max_temp = response["main"]["temp_max"]
        city_min_temp = response["main"]["temp_min"]
        city_pressure = response["main"]["pressure"]
        city_humidity = response["main"]["humidity"]
        city_clouds = response["clouds"]["all"]
        city_wind = response["wind"]["speed"]
        city_country = response["sys"]["country"]
        city_weather_description = response["weather"][0]["description"]
        city_date = response["dt"]

        # Append the City information into city_data list
        city_data.append({
            "City": city, 
            "Lat": city_lat, 
            "Lng": city_lng,
            "Temp": city_temp,
            "Max Temp": city_max_temp,
            "Min Temp": city_min_temp,
            "Pressure": city_pressure,
            "Humidity": city_humidity,
            "Cloudiness": city_clouds,
            "Wind Speed": city_wind,
            "Country": city_country,
            "Description": city_weather_description,
            "Date": city_date})
        

```

```python
city_df = pd.DataFrame(city_data)
```
